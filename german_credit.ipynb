# Problem 2: German Credit Dataset

Load the CSV file into a Pandas DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("GermanCredit.csv")
df.head()
```

## Preprocessing

Drop the 3 columns with the highest number of 'none' values.

```python
none_counts = (df == 'none').sum()
cols_to_drop = none_counts.sort_values(ascending=False).head(3).index
df = df.drop(columns=cols_to_drop)
```

Remove unnecessary apostrophes from values.

```python
df = df.replace("'", "", regex=True)
```

Fix category values for checking_status, savings_status, class, and employment.

```python
df['checking_status'] = df['checking_status'].replace({
    'no checking': 'No Checking',
    '<0': 'Low',
    '0<=X<200': 'Medium',
    '>=200': 'High'
})

df['savings_status'] = df['savings_status'].replace({
    'no known savings': 'No Savings',
    '<100': 'Low',
    '100<=X<500': 'Medium',
    '500<=X<1000': 'High',
    '>=1000': 'High'
})

df['class'] = df['class'].replace({
    'good': 1,
    'bad': 0
})

df['employment'] = df['employment'].replace({
    'unemployed': 'Unemployed',
    '<1': 'Amateur',
    '1<=X<4': 'Professional',
    '4<=X<7': 'Experienced',
    '>=7': 'Expert'
})
```

## Analysis

Count of foreign workers for each class.

```python
pd.crosstab(df['foreign_worker'], df['class'])
```

Count of employment categories for each savings_status.

```python
pd.crosstab(df['employment'], df['savings_status'])
```

Average credit_amount for single males with 4<=X<7 years employment.

```python
df[(df['personal_status'] == 'male single') & 
   (df['employment'] == 'Experienced')]['credit_amount'].mean()
```

Average duration for each job type.

```python
df.groupby('job')['duration'].mean()
```

Most common checking_status and savings_status for education.

```python
edu = df[df['purpose'] == 'education']
print("Most common checking status:", edu['checking_status'].mode()[0])
print("Most common savings status:", edu['savings_status'].mode()[0])
```

## Visualization

Bar chart of savings_status vs personal_status.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

pd.crosstab(df['savings_status'], df['personal_status']).plot(kind='bar', ax=axes[0])
axes[0].set_title("Savings Status vs Personal Status")
axes[0].set_xlabel("Savings Status")
axes[0].set_ylabel("Count")

pd.crosstab(df['checking_status'], df['personal_status']).plot(kind='bar', ax=axes[1])
axes[1].set_title("Checking Status vs Personal Status")
axes[1].set_xlabel("Checking Status")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

Bar graph for people with credit_amount more than 4000, mapping property_magnitude to average age.

```python
high_credit = df[df['credit_amount'] > 4000]

high_credit.groupby('property_magnitude')['age'].mean().plot(kind='bar')
plt.title("Average Age by Property Magnitude (Credit Amount > 4000)")
plt.xlabel("Property Magnitude")
plt.ylabel("Average Age")
plt.tight_layout()
plt.show()
```

Pie charts for people with High savings_status and age above 40.

```python
filtered = df[(df['savings_status'] == 'High') & (df['age'] > 40)]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

filtered['personal_status'].value_counts().plot(kind='pie', ax=axes[0], autopct='%1.1f%%')
axes[0].set_title("Personal Status")
axes[0].set_ylabel("")

filtered['credit_history'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%')
axes[1].set_title("Credit History")
axes[1].set_ylabel("")

filtered['job'].value_counts().plot(kind='pie', ax=axes[2], autopct='%1.1f%%')
axes[2].set_title("Job")
axes[2].set_ylabel("")

plt.tight_layout()
plt.show()
```
